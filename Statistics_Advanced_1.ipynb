    A random variable does not represent randomness itself; instead, it represents numerical outcomes associated with random events.

    Importance of Random Variables:

    They help quantify uncertainty

    They are essential in statistics, data analysis, and machine learning

2. What are the types of random variables?
     - Random variables are mainly classified into two types based on the nature of values they can take:

     a. Discrete Random Variable

    A discrete random variable takes countable values, usually integers.

    Examples:

    Number of students in a class

    Number of heads in 5 coin tosses

    Number of defective items in a batch

    Key Characteristics:

    Takes finite or countably infinite values

    Described using a probability mass function (PMF)


    b. Continuous Random Variable

    A continuous random variable takes any value within a given range.

    Examples:

    Height of a person

    Weight of an object

    Time taken to complete a task

    Key Characteristics:

    Takes infinite uncountable values

    Described using a probability density function

6. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
      - The Central Limit Theorem (CLT) states that the distribution of sample means approaches a normal distribution as the sample size increases, regardless of the population's original distribution, provided the samples are independent and identically distributed.

     -> Key Points

        Applies to large sample sizes (typically n>=30)

        Works even if the population distribution is not normal

        Mean of sample means equals population mean

        Standard deviation decreases with larger samples

     -> Why CLT is Critical

        Enables use of normal distribution in real-world problems

        Forms the foundation of hypothesis testing

        Allows estimation of population parameters

        Justifies many statistical methods

     -> Practical Example

        Estimating average customer spending

         Survey analysis

        Quality testing in manufacturing

7.  What is the significance of confidence intervals in statistical analysis?
     - A confidence interval (CI) is a range of values within which a population parameter (such as mean or proportion) is expected to lie with a certain level of confidence (commonly 90%, 95%, or 99%).

     -> Significance of Confidence Intervals

        They provide a measure of uncertainty around an estimate.

        Instead of giving a single value, they give a range, making results more informative.

        Higher confidence levels indicate greater reliability, though with wider intervals.

        They are widely used in hypothesis testing, survey analysis, and decision-making.

    -> Example
      
      If a 95% confidence interval for average income is (‚Çπ40,000, ‚Çπ45,000), it means we are 95% confident that the true population mean lies within this range.

```python
# 9.  Write a Python program to generate 1000 random numbers from a normal
# distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
# deviation using NumPy, and draw a histogram to visualize the distribution.
# (Include your Python code and output in the code box below.)


#solution:-
# #A normal distribution is defined by its mean and standard deviation.
# Using NumPy, we can generate random samples, compute summary statistics, and visualize the distribution using a histogram.


import numpy as np
import matplotlib.pyplot as plt

# Generate random numbers
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute statistics
mean_value = np.mean(data)
std_value = np.std(data)

print("Mean:", mean_value)
print("Standard Deviation:", std_value)

# Plot histogram
plt.hist(data, bins=30, edgecolor='black')
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram of Normal Distribution (Mean=50, Std=5)")
plt.show()
```

```python
#10.  You are working as a data analyst for a retail company. The company has
# collected daily sales data for 2 years and wants you to identify the overall sales trend.
# daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
#  235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
# ‚óè Explain how you would apply the Central Limit Theorem to estimate the average sales
# with a 95% confidence interval.
# ‚óè Write the Python code to compute the mean sales and its confidence interval.
# (Include your Python code and output in the code box below.)


#solution:-

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Explanation: Applying the Central Limit Theorem (CLT)

# The Central Limit Theorem (CLT) states that for a sufficiently large sample size,
# the sampling distribution of the sample mean approaches a normal distribution, regardless of the original data distribution.

# In this case:
# Daily sales data represents a sample from the population of all possible sales days
# Sample size is reasonably large
# Therefore, CLT allows us to:
# Assume normality of the sample mean
# Estimate the average sales
# Construct a 95% confidence interval for the true mean sales
# 95% Confidence Interval Formula


# CI = x^- +- z * s/n




import numpy as np
import math

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Sample size
n = len(daily_sales)

# Mean and standard deviation
mean_sales = np.mean(daily_sales)
std_dev = np.std(daily_sales, ddof=1)   # sample standard deviation

# Z-value for 95% confidence
z = 1.96

# Margin of error
margin_error = z * (std_dev / math.sqrt(n))

# Confidence interval
lower_bound = mean_sales - margin_error
upper_bound = mean_sales + margin_error

print("Mean Daily Sales:", mean_sales)
print("95% Confidence Interval:", (lower_bound, upper_bound))
```
